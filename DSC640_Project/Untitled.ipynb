```python
import pandas as pd
import numpy as np
```

```python
#reading in motor vehicles data to add/change columns to be more useable in Tableau
df = pd.read_csv("motor-vehicle-crash-reports-1.csv")
df.head()
```

```python
severity = []
df.reset_index(drop=True)
for i in range(len(df)):
    if not np.isnan(df.loc[i].Severity1) :
        severity.append("Severe")
    elif not np.isnan(df.loc[i].Severity2):
        severyity.append("Severe")
    elif not np.isnan(df.loc[i].Severity3):
        severyity.append("Severe")
    elif not np.isnan(df.loc[i].Severity4):
        severyity.append("Severe")
    elif not np.isnan(df.loc[i].Severity5):
        severyity.append("Severe")
    else:
        severity.append("Not Severe")
        
df['Severity'] = severity
```

```python
df.Severity.value_counts()
```

```python
df.to_csv("MotorVehicles.csv")
```

```python
#reading in data set to change it
df = pd.read_csv("Airplane_Crashes_and_Fatalities_Since_1908.csv")
df.head()
```

```python
operator = df.Operator
op =[]
m = "Military"
p = "Private"
for i in range(len(operator)):
    print(type(operator[i]))
    print(operator[i])
    try:
        if operator[i].find(m) != -1:
            op.append("Military")
        elif operator[i].find(p) != -1:
            op.append("Private")
        else:
            op.append("Public")
    except:
        op.append("Unknown")
```

```python
df["OpType"] = op
```

```python
df.to_csv("AirplaneCrashesEdited.csv")
```
