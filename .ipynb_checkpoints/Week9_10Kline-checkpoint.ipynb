```python
#read files
costco = pd.read_csv("C:/Users/Matt Kline/Documents/GitHub/DSC640/ex5-2/costcos-geocoded.csv")
ppg = pd.read_csv("C:/Users/Matt Kline/Documents/GitHub/DSC640/ex5-2/ppg2008.csv")
```

```python
#contour map
x = ppg.FGP
y = ppg.FTP
x1,y1 = np.meshgrid(x,y)
z = np.sin(x1) * np.cos(y1)

cs = plt.contour(x,y,z)
#plt.clabel(cs, inline=1, fontsize=5)
plt.xticks([.35,.4,.45,.5,.55,.6,.65])
plt.yticks([.55,.6,.65,.7,.75,.8,.85,.9,.95])
plt.yticks
plt.xlabel("Field Goal Precentage")
plt.ylabel("Free Throw Precentage")
plt.title("Contour Plot for FPG and FTP")
```

```python
#heat map
sns.heatmap(ppg.corr(), annot=False)
plt.show()
```

```python
from mpl_toolkits.basemap import Basemap
```
