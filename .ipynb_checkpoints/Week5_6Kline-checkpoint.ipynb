```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import squarify
```

```python
#read file
df = pd.read_csv("C:/Users/Matt Kline/Documents/GitHub/DSC640/ex3-3/unemployement-rate-1948-2010.csv")
df.head()
```

```python
#get the unique labels for periods, and total value for Value at each label to create tree map
labels = df.Period.unique()
sums = []
for i in range(len(labels)):
    sums.append(sum(df[df['Period'] == labels[i]]['Value']))

print(sums)
print(labels)
```

```python
#tree map
squarify.plot(sizes=sums,label=labels)
plt.title("Treemap of Value totals per label")
plt.show()
```

```python
#area chart
plt.fill_between(df.Year,df.Value)
plt.title("Area chart showing value over year")
plt.xlabel("Year")
plt.ylabel("Value")
plt.show()
```

```python
#df pivot
piv_df = df.pivot(index='Year', columns='Period', values='Value')
piv_df = piv_df.fillna(0.0)
piv_df
```

```python
#stacked area chart
piv_df.plot.area()
plt.title("Area chart showing value over year split by Period")
plt.xlabel("Year")
plt.ylabel("Value")
plt.legend(bbox_to_anchor = (1,1))
plt.show()
```
