```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#read file
df = pd.read_csv("C:/Users/Matt Kline/Documents/GitHub/DSC640/ex4-2/crimerates-by-state-2005.csv")
#drop united states
df=df[df["state"] != 'United States']
df.reset_index(drop=True,inplace=True)
df.head()
```

```python
#scatter plot
plt.scatter(df.burglary, df.aggravated_assault)
plt.title("Burglary vs Assault")
plt.ylabel("Aggravated Assault")
plt.xlabel("Burglary")
plt.show()
```

```python
#bubble chart
sns.scatterplot(data=df, x="population", y="larceny_theft", size="population", legend=False, sizes=(100, 2000))
plt.title("Population vs Larceny Theft")
plt.ylabel("Larceny Theft")
plt.xlabel("Population")
plt.show()
```

```python
#density map
sns.kdeplot(x=df["population"],data=df, hue=df["state"])
plt.xlabel("Aggravated Assualt")
plt.ylabel("Distribution")
plt.title("Analysis of Aggravated Assualt across the United States")
plt.show()
```
